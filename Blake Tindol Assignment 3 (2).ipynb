```python
#3. The first name of how many presidents starts with the letter 'B' ? (20 points)
file = open("USPres.txt",'r')
Counter = 0
for line in file:
    line = line.rstrip()
    line = line.split(" ")[0]
    if line[0] == "B":
        Counter = Counter + 1
    
print("There are " + str(Counter) + " presidents with the first name starting with B.")
```
