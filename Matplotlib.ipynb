```python
# Import the Matplotlib

import matplotlib as mlp

# import pyplot module / collection of functions as plt

import matplotlib.pyplot as plt

# The following command allows to see plots in the Jupyter library in can you can not see them.

%matplotlib inline

# plt module has a plot function to make plots from data 

plt.plot([5,7,2,4])

plt.show() # Opens a new interactive window to show the second plot 

plt.plot([5,7,2,4], [4,6,9,2], 'ro') # ro stands for round objects by default a line is plotted 

# Plotting two numeric lists 

year = [2004,2005,2006,2007]
unemployment_rate = [2,5,3,1]

plt.show() 

plt.plot(year,unemployment_rate)
```

```python
# Understanding some of the parameters of the plot function
# Example 2 : Changing the width, color and complexion of the plots 

import matplotlib as mlp

import matplotlib.pyplot as plt

%matplotlib inline

year = [2004,2005,2006,2007]
unemployment_rate = [2,5,3,1]

# Parameter linewidth 

plt.plot(year,unemployment_rate, linewidth = 4)

plt.show()

# Parameter for line style 

plt.plot(year,unemployment_rate,linestyle = 'dashed', linewidth = 4)

# check this url for all supported linestyles 

# https://matplotlib.org/3.1.0/gallery/lines_bars_and_markers/linestyles.html

# Adding marker to the plots 

plt.show()

plt.plot(year,unemployment_rate,linestyle = 'solid', linewidth = 4, marker = '1', markersize = 10)

# Refer to this url for more marker options 

# https://matplotlib.org/3.1.0/gallery/lines_bars_and_markers/marker_reference.html

plt.show()

# Various types of markers, line and colors 

plt.plot(year,unemployment_rate,'b-')

# Refer to the following url for more information 

# https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.plot.html
```

```python
# Adding title and axis labels to the plots 
# Example 3 - Customizing Plots 

import matplotlib as mlp

import matplotlib.pyplot as plt

year = [2004,2005,2006,2007]
unemployment_rate = [2,5,3,1]

plt.plot(year,unemployment_rate, linewidth = 4)

# Adding Title to the Plot 
plt.title("Employment Rate Over the Years")

# Adding Labels to the x-axis 
plt.xlabel("YEAR")

# Adding labels to the y-axis 

plt.ylabel("EMPLOYMENT RATE")

#help(plt.ylabel)

# Changing the color of the labels 

plt.show()

plt.plot(year,unemployment_rate, 'b*')

# Adding Title to the Plot 
plt.title("Employment Rate Over the Years", color = "purple")

# Adding Labels to the x-axis 
plt.xlabel("YEAR", color = "red")

# Adding labels to the y-axis 

plt.ylabel("EMPLOYMENT RATE", color = "green")



```

```python
# Example 4 Customizing multiple plots in the same window 

import matplotlib as mlp

import matplotlib.pyplot as plt

year = [2004,2005,2006,2007]
unemployment_rate = [2,5,3,1]
inflation = [4,3,5,6]

# Lets plot unemployment rate and inflation in the same plot 

plt.plot(year,unemployment_rate,'r*',markersize = 15)

plt.plot(year,inflation,'bo',markersize = 15)

plt.title("Employment Rate & Inflation vs Year", color = "purple")

# Adding legend to the plot 

plt.legend(['Unemployment_rate','Inflation','Third'],loc=0) #loc changes the position of legend

# For possible values of legend check the following url

# https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.legend.html
    
```

```python
# Creating subchars using subplot()

# Determine how many subplot you want in one frame 

# plt.subplot('rows','columns','index')

year = [2004,2005,2006,2007]
unemployment_rate = [2,5,3,1]
inflation = [4,3,5,6]
gdp_growth = [2,3,1,2]
stock_growth = [8,5,4,8]

plt.subplot(2,2,1)
plt.plot(year ,unemployment_rate,'b*')
plt.title("Unemployment_rate vs Year", color = "purple")


plt.subplot(2,2,2)
plt.plot(year,inflation,'g^')
plt.title("Inflation vs Year", color = "purple")


plt.subplot(2,2,3)
plt.plot(year,gdp_growth,'ro')
plt.title("Gdp_growth vs Year", color = "purple")


plt.subplot(2,2,4)
plt.plot(year,stock_growth,'b')
plt.title("Stock_Growth vs Year", color = "purple")

# plt.tight_layout() Use this to remove the merging of xlabels and titles


```

```python
# Other plots in Python 

company = ['A','B','C','D']
share = [20,30,40,10]

plt.pie(share); # See what removing the semicolon does 

plt.show()
plt.pie(share, labels = company); # See what removing the semicolon does 
# Explode the largest slice and add % in the plots

plt.show()
explode = (0, 0, 0.1, 0)

plt.pie(share, labels = company, explode = explode,autopct='%1.1f%%');

# For more information on the pie charts check the link below 


# https://matplotlib.org/3.2.1/gallery/pie_and_polar_charts/pie_features.html#sphx-glr-gallery-pie-and-polar-charts-pie-features-py

```

```python
# Plotting a bar plot from the dataframe values 

import pandas as pd

import numpy as np # This library is imported to use np.random.rand()

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'


# Creating a Pandas dataframe with 4 columns and 10 rows 

df1 = pd.DataFrame(np.random.rand(10, 4), columns = ['var1', 'var2', 'var3', 'var4'])

df1

df1.plot(kind = "bar")

# Stacked bar plot 

df1.plot(kind = "bar", stacked = True)

# A Histogram of all columns in the dataframe 

df1.hist()

# Plotting the relationship between var1 (x axis) and var2 (y axis)

plt.show()


plt.plot(df1.var1,df1.var2, 'b*', markersize = 15)

# Adding Title to the Plot 
plt.title("Var1 vs Var2", color = "purple")

# Adding Labels to the x-axis 
plt.xlabel("Var1", color = "red")

# Adding labels to the y-axis 

plt.ylabel("Var2", color = "green")

```
